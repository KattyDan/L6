```python
import numpy as np #библиотка для работы с массивами и мат.опер

import matplotlib.pyplot as plt


# Генерирауия массива x от 0 до 10 (100 точек)

x = np.linspace(0, 10, 100)


# Вычисляем sin(x) и cos(x)

y = np.sin(x)

z = np.cos(x)


# Создане графика

plt.figure(figsize=(8, 5))  # Размер графика

plt.plot(x, y, label='sin(x)', color='blue')  # График синуса

plt.plot(x, z, label='cos(x)', color='red')   # График косинуса


# Добавление подписей

plt.title('Графики sin(x) и cos(x)')  # настройка графика

plt.xlabel('Ось X')                  # добавление заголовка

plt.ylabel('Ось Y')

plt.legend()                         # Легенда с метками синуса и кос

plt.grid(True)                       # Сетка

plt.show()  # Показать график
```

```python
import numpy as np



# Создаем матрицу 5x5 со случайными числами от 1 до 10

matrix = np.random.randint(1, 11, size=(5, 5))  # от 1 до 10 (11 не входит)



# Вычисляем характеристики

mean = np.mean(matrix)  # Среднее

max_val = np.max(matrix)  # Максимум

min_val = np.min(matrix)  # Минимум

col_sums = np.sum(matrix, axis=0)  # Сумма по столбцам



print("Матрица:\n", matrix) #ввод маьрицы ,перенос строки перед выводом матрицы

print("\nСреднее:", mean) #вычисление и ввод ср.знач, сред. всех ариф. чисел

print("Максимум:", max_val)

print("Минимум:", min_val)

print("Сумма по столбцам:", col_sums) #ввчисление и вывод из столбцов, сум. вдоль всех столб.
```

```python

import matplotlib.pyplot as plt # импорт библиотеки для рисования графиков

# список оценок студентов (баллы)
scores = [60, 70, 75, 80, 85, 90, 95, 100, 100, 85, 75, 70, 90, 80, 65, 60, 95, 85]

#  список групп, к которым относятся студенты (та же длина, что и у scores)
groups = [
    'A', 'A', 'A', 'A', 'A', 'A',   # первые 6 студентов из группы A
    'B', 'B', 'B', 'B', 'B', 'B',   # следующие 6 — группа B
    'C', 'C', 'C', 'C', 'C', 'C'   # последние 6 — группа С
   ]
    #  Гистограмма баллов

plt.hist(scores, bins=5, color='orange', edgecolor='red') #hist строим гистограмму
plt.title("Гистограмма баллов")         # Заголовок
plt.xlabel("Балл")                      # Подпись оси X
plt.ylabel("Количество студентов")      # Подпись оси Y
plt.show()                               # Показать график


# создание словаря, чтобы собрать баллы по каждой группе
group_scores = {}

# сложение баллов по группам
for score, group in zip(scores, groups):
    if group not in group_scores:
        group_scores[group] = []        # если группа новая — создаем список
    group_scores[group].append(score)   # добавляем балл в список группы

# считаем ср.значение для каждой группы
average_scores = {}
for group in group_scores:
    average_scores[group] = sum(group_scores[group]) / len(group_scores[group]) #bar столбчатая диагр.

# строим столбчатую диаграмму
plt.bar(average_scores.keys(), average_scores.values(), color='yellow')
plt.title("Средний балл по группам")
plt.xlabel("Группа")
plt.ylabel("Средний балл")
plt.show()
```
