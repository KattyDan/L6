```python
import numpy as np #      .

import matplotlib.pyplot as plt


#   x  0  10 (100 )

x = np.linspace(0, 10, 100)


#  sin(x)  cos(x)

y = np.sin(x)

z = np.cos(x)


#  

plt.figure(figsize=(8, 5))  #  

plt.plot(x, y, label='sin(x)', color='blue')  #  

plt.plot(x, z, label='cos(x)', color='red')   #  


#  

plt.title(' sin(x)  cos(x)')  #  

plt.xlabel(' X')                  #  

plt.ylabel(' Y')

plt.legend()                         #      

plt.grid(True)                       # 

plt.show()  #  
```

```python

import matplotlib.pyplot as plt #     

#    ()
scores = [60, 70, 75, 80, 85, 90, 95, 100, 100, 85, 75, 70, 90, 80, 65, 60, 95, 85]

#   ,     (  ,    scores)
groups = [
    'A', 'A', 'A', 'A', 'A', 'A',   #  6    A
    'B', 'B', 'B', 'B', 'B', 'B',   #  6   B
    'C', 'C', 'C', 'C', 'C', 'C'   #  6   
   ]
    #   

plt.hist(scores, bins=5, color='orange', edgecolor='red') #hist  
plt.title(" ")         # 
plt.xlabel("")                      #   X
plt.ylabel(" ")      #   Y
plt.show()                               #  


#  ,      
group_scores = {}

#    
for score, group in zip(scores, groups):
    if group not in group_scores:
        group_scores[group] = []        #      
    group_scores[group].append(score)   #     

#  .   
average_scores = {}
for group in group_scores:
    average_scores[group] = sum(group_scores[group]) / len(group_scores[group]) #bar  .

#   
plt.bar(average_scores.keys(), average_scores.values(), color='yellow')
plt.title("   ")
plt.xlabel("")
plt.ylabel(" ")
plt.show()
```
